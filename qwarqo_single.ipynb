```python
import requests
from bs4 import BeautifulSoup

# Define the URL of the website to scrape
url = 'https://qwarqo.com/seemore/Single%20Room'

# Send a GET request to the website
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Create a BeautifulSoup object from the response content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the container that holds the property listings
    property_listings = soup.find_all('div', class_='col-6 col-sm-6 col-md-3 col-lg-2 col-xxl-2')

    # Set a counter variable to limit the number of properties to extract
    count = 0

    # Iterate over each property listing and extract relevant information
    for listing in property_listings:
        # Check if the desired number of properties has been extracted
        if count >= 10:
            break

        # Extract the property title
        title = listing.find('h6', class_='card-title m-1').text

        # Extract the property description
        description = listing.find('div', class_='card-body p-2').text

        # Extract the property price
        #price = listing.find('span', class_='price').text

        # Extract the property location
        #location = listing.find('span', class_='card-text text-muted').text

        # Print the extracted information
        print('Title:', title)
        print('Description:', description)
        #print('Price:', price)
       # print('Location:', location)
        print('-------------------')

        # Increment the counter
        count += 1
else:
    print('Failed to retrieve the website. Status code:', response.status_code)
```
